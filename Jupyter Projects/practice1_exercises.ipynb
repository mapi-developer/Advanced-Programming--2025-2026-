
## Exercise 1 

Write a function to sum the squares of the values of a list, once with and once without list comprehension. Use the %%timeit function to test the speed of the two implementations on a randomly generated list of number of length 1000.

```python
l = [1, 2, 3, 4, 5]

def sqr_sum(l: list) -> int:
    answ = 0
    for i in l:
        answ += i**2
    return answ

def sqr_sum_compr(l: list) -> int:
    return sum([i**2 for i in l])

print(sqr_sum_compr(l))
```

## Assignment/exercise 2 (5 minutes)

Write a function to return the index of first occurrence of an item in a list. 

```python
def idx(item: int, l: list) -> int | bool:
    for i, num in enumerate(l):
        if num == item: return i
    return False

print(idx(1, [2, 5, 4, 1]))
```

## Assignment/exercise 3 (5 minutes)

Extend previous function (#2) to find the nth occurrence of an item in a list. 

```python
def idx(item: int, l: list, all_idxs: bool = False) -> int | bool | list:
    idxs = [i for i, num in enumerate(l) if num == item]
    if idxs and all_idxs:
        return idxs
    elif idxs:
        return idxs[0]
    else:
        return False

print(idx(1, [2, 1, 5, 4, 1, 5, 1], True))
```

## Assignment/exercise 5 (3 minutes)

Write two functions to test different kinds of equality between two lists: item-wise vs. object equality. Item-wise equality means that two lists have the same elements in the same order. Object equality means that the two variable names are referring to the same list in memory.

```python
def item_check(l1: list, l2: list) -> bool:
    return l1 == l2

def object_check(l1: list, l2: list) -> bool:
    return l1 is l2
```

Write a function to determine whether given string is a palindrome. Try to improve the speed of your code and check the results using %timeit.

```python
def is_palindrome(s: str) -> bool:
    return s.lower() == s[::-1].lower()
```

## Assignment/exercise 7 (15 minutes)

Write a function to find given substring in a string using naive (brute force) pattern matching.

```python
def find_match(pattern: str, s: str) -> bool:
    length_of_pattern = len(pattern)
    for i in range(len(s)):
        if s[i:length_of_pattern+i] == pattern:
            return True
    return False
```

## Assignment/exercise 8 (15 minutes)

Given a list having both positive and negative integers, find the largest sublist with 0 sum. (Source: Largest subarray with 0 sum | Practice | GeeksforGeeks)

```python
def largest_zero_sum_sublist(arr: list[int]) -> list[int]:
    # This dictionary will store {prefix_sum: index_where_it_first_occurred}
    prefix_map = {}
    current_sum = 0
    max_length = 0
    
    # Trackers to help us slice the final list
    start_index = 0
    end_index = 0

    for i, num in enumerate(arr):
        current_sum += num

        # Case 1: The sublist starts from the very beginning (index 0)
        if current_sum == 0:
            max_length = i + 1
            start_index = 0
            end_index = i + 1

        # Case 2: We've seen this exact sum before
        elif current_sum in prefix_map:
            # Calculate the length of this 0-sum sublist
            current_length = i - prefix_map[current_sum]
            
            # If it's the largest we've seen, update our trackers
            if current_length > max_length:
                max_length = current_length
                # The sublist starts right after the first time we saw this sum
                start_index = prefix_map[current_sum] + 1
                end_index = i + 1
                
        # Case 3: We haven't seen this sum before, so record its first appearance
        else:
            prefix_map[current_sum] = i

    # Return the largest sublist (or an empty list if none found)
    return arr[start_index:end_index] if max_length > 0 else []
```

## Assignment/exercise 9 (20 minutes)

Given a collection of integer intervals (e.g.: [(1,5), (2,3), (6,8),(8,9)]), merge all the overlapping intervals. Result should be a list of tuples. In this case: [(1,5),(6.9)]

```python
def merge_intervals(intervals: list[tuple]) -> list[tuple]:
    # Edge case: return empty list if input is empty
    if not intervals:
        return []

    # 1. Sort the intervals based on their starting value
    intervals.sort(key=lambda x: x[0])

    # Initialize the merged list with the first interval
    merged = [intervals[0]]

    # 2. Iterate through the rest of the intervals
    for current in intervals[1:]:
        last_merged = merged[-1] # Look at the most recently merged interval

        # If the current interval overlaps with the last merged one
        if current[0] <= last_merged[1]:
            # Update the end point to be the maximum of both ends
            # Using max() handles cases like completely swallowed intervals e.g., (1,5) and (2,3)
            merged[-1] = (last_merged[0], max(last_merged[1], current[1]))
        else:
            # No overlap, so we add it as a new distinct interval
            merged.append(current)

    return merged
```
