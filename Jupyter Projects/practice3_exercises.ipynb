### Exercise 1: Singly Linked List

In this exercise you will implement a **singly linked list** from scratch — one of the most fundamental data structures in computer science.

Unlike Python lists (which are backed by dynamic arrays), a linked list stores elements as individual **nodes** scattered in memory, each pointing to the next. This makes insertions and deletions $O(1)$ at a known position (no shifting required), but random access $O(n)$ (you must traverse from the head).

**Step 1 — The `Node` class:**

Start by creating a class called `Node`. Each instance represents one element of the linked list and should contain:
- `data` — the value stored in this node.
- `next` — a reference (pointer) to the next `Node` in the list, initialized to `None`.

```
Node("A") → Node("B") → Node("C") → None
```

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
````

**Step 2 — The `LinkedList` class:**

Now implement the `LinkedList` class itself. It should have an `__init__` method that creates an empty list by setting `self.head = None`.

Implement the following methods:

| Method | Description | Key idea |
|---|---|---|
| `append(data)` | Add a new node at the **end** of the list. | Traverse to the last node (where `.next is None`), then link the new node there. |
| `insert(prev_node_data, data)` | Insert a new node **after** the node containing `prev_node_data`. | Find the target node, then rewire: `new_node.next = target.next`, then `target.next = new_node`. |
| `remove(data)` | Remove the first node containing `data`. | Keep track of the **previous** node so you can bridge the gap: `prev.next = current.next`. |
| `find(data)` | Return `True` if any node contains `data`, `False` otherwise. | Simple traversal with an equality check. |
| `show_items()` | Print all elements in order. | Traverse from head to tail, printing each node's data. |

**Test your implementation:**
```python
ll = LinkedList()
ll.append(1); ll.append(2); ll.append(3)
ll.show_items()        # → 1 2 3
ll.insert(2, 2.5)
ll.show_items()        # → 1 2 2.5 3
ll.remove(2)
ll.show_items()        # → 1 2.5 3
print(ll.find(2.5))   # → True
print(ll.find(2))     # → False
```

````python
class LinkedList:
    class Node:
        def __init__(self, data):
            self.data = data
            self.next = None

        def __str__(self):
            return f"Value: {self.data}, Next: {type(self.next)}\n"

    def __init__(self):
        self.head = None

    def __str__(self):
        s = "forward: "
        current = self.head
        while current is not None:
            s = s + f"{current.data}, "
            current = current.next
        return s + "\n...\n"

    def append(self, data):
        node_to_append = self.Node(data)
        if self.head is None:
            self.head = node_to_append
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = node_to_append

    def insert(self, prev_node_data, data):
        current = self.head
        if current is None:
            self.append(data)
        while current != None and current.data != prev_node_data:  
            current = current.next
        if current == None: return
        next_node = current.next
        current.next = self.Node(data)
        current.next.next = next_node

    def remove(self, data):
        current = self.head
        if current is None: return
        if current.data == data:
            current = current.next
        while current.next != None and current.next.data != data:
            current = current.next
        if current is None: return
        current.next = current.next.next

    def find(self, data):
        current = self.head
        while current != None and current.data != data:
            current = current.next
        else:
            return True
        if current == None: return False
        return True

    def show_items(self):
        print(self)
````

````python
# Test your implementation
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.show_items()        # → 1 2 3
ll.insert(2, 2.5)
ll.show_items()        # → 1 2 2.5 3
ll.remove(2)
ll.show_items()        # → 1 2.5 3
print(ll.find(2.5))    # → True
print(ll.find(3))      # → False
````

### Exercise 2: FIFO Queue — Two Implementations

Implement a **FIFO (First-In, First-Out) queue** in two different ways and compare their performance.

A queue is like a line at a store: the first person to join is the first to be served. The two core operations are **enqueue** (join the back of the line) and **dequeue** (leave from the front).

**Implementation A — `ListQueue`** (using a plain Python `list`):
- Use `.append(item)` to enqueue.
- Use `.pop(0)` to dequeue from the front.

**Implementation B — `DequeQueue`** (using `collections.deque`):
- Use `.append(item)` to enqueue.
- Use `.popleft()` to dequeue from the front.

Both classes should support these methods:

| Method | Description |
|---|---|
| `__init__()` | Initialize an empty queue. |
| `enqueue(item)` | Add an item to the back of the queue. |
| `dequeue()` | Remove and return the item at the front. |
| `peek()` | Return the front item **without** removing it. |
| `size()` | Return the number of items in the queue. |

**Think about:** Why might these two implementations differ in performance? What is the time complexity of `.pop(0)` on a list vs. `.popleft()` on a deque?

````python
class ListQueue(list):
    def __init__(self):
        super().__init__()

    def enqueue(self, item):
        self.append(item)

    def dequeue(self):
        self.pop(0)

    def peek(self):
        return self[0]

    def size(self):
        return len(self)



lq = ListQueue()
for i in range(10):
    lq.enqueue(i)

for i in range (4):
    lq.dequeue()

print(lq)
````

````python
from collections import deque

class DequeQueue(deque):
    def __init__(self):
        super().__init__()

    def enqueue(self, item):
        self.append(item)

    def dequeue(self):
        self.popleft()

    def peek(self):
        return self[0]

    def size(self):
        return len(self)


lq = ListQueue()
for i in range(10):
    lq.enqueue(i)

for i in range (4):
    lq.dequeue()

print(lq)
````

### Exercise 2b: Performance Comparison

Compare the two queue implementations by **enqueuing** 100,000 elements and then **dequeuing** all of them. Measure the total time using the `time` module.

Which implementation is faster, and by how much? The result illustrates why choosing the right data structure matters: `list.pop(0)` is $O(n)$ (every remaining element must shift left), while `deque.popleft()` is $O(1)$ (doubly-linked list, no shifting). Over 100,000 dequeue operations, this difference compounds dramatically.

````python
import time

def measure_performance(queue_class, operations=100_000):
    start_time = time.time()
    ql = queue_class()
    for _ in range(operations):
        ql.enqueue(time.time())
    for _ in range(operations):
        ql.dequeue()
    end_time = time.time()
    return end_time - start_time

list_queue_time = measure_performance(ListQueue)
deque_queue_time = measure_performance(DequeQueue)
print(f"ListQueue: {list_queue_time:.4f}s")
print(f"DequeQueue: {deque_queue_time:.4f}s")
````

### Exercise 3: Browser History with a Stack (LIFO)

Implement a **web browser's back-button history** using a **stack** (Last-In, First-Out).

When you visit a new page, it gets pushed onto the stack. When you press "Back", the most recent page is popped off, revealing the previous one — exactly how a stack works.

Implement a class `BrowserHistory` with the following methods:

| Method | Description |
|---|---|
| `__init__()` | Initialize an empty history stack (use a Python list). |
| `visit(url)` | **Push** a URL onto the stack (equivalent to navigating to a new page). |
| `go_back()` | **Pop** and return the most recent URL. If the history is empty, return an appropriate message instead of crashing. |
| `current_page()` | Return the URL on top of the stack (the page you're currently viewing) **without** removing it. |
| `show_history()` | Return or print the entire history stack. |

**Test your implementation:**
```python
bh = BrowserHistory()
bh.visit("google.com")
bh.visit("github.com")
bh.visit("claude.ai")
print(bh.current_page())   # → claude.ai
print(bh.go_back())        # → claude.ai  (popped)
print(bh.current_page())   # → github.com
print(bh.show_history())   # → ['google.com', 'github.com']
```

**Think about:** How is this different from a queue? What real-world browser features would require a *second* stack (hint: the "Forward" button)?

````python
class BrowserHistory(list):
    def __init__(self):
        super().__init__()

    def visit(self, url):
        if not url: return
        self.append(url)

    def go_back(self):
        if self == []: return None
        return f"{self.pop()} (popped)"

    def current_page(self):
        if self == []: return None
        return self[-1]

    def show_history(self):
        return self
````

````python
# Test your implementation
browser_history = BrowserHistory()

browser_history.visit("page1.com")
browser_history.visit("page2.com")
browser_history.visit("page3.com")

print("Current page:", browser_history.current_page())
print("Going back to:", browser_history.go_back())
print("Current page after going back:", browser_history.current_page())
print("Browser history:", browser_history.show_history())
````
