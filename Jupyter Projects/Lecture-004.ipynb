## 01 - Bubble sort

```python
def bubble_sort(li):
    n = len(li)
    num_comparisons, num_swaps = 0, 0
    for topinx in range(n-1, 0, -1):
        for inx in range(topinx):
            if li[inx] > li[topinx]:
                li[inx], li[topinx] = li[topinx], li[inx]
                num_swaps += 1
            num_comparisons += 1
    print(f"Number of comparisons: {num_comparisons}; number of swaps: {num_swaps}")
```

```python
## test
##copied_list = li_master.copy()
copied_list = [item for item in li_master]
bubble_sort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```

## Comb sort

```python
def get_next_gap(gap):
    gap = (gap * 10) // 13
    return gap if gap > 1 else 1

def comb_sort(li):
    n = len(li)
    gap = n
    num_comparisons, num_swaps = 0, 0
    while gap > 1:
        gap = get_next_gap(gap)
        print(f"next gap is: {gap}")
        for i in range(n-gap):
            if li[i] > li[i+gap]:
                li[i], li[i+gap] = li[i+gap], li[i]
                num_swaps += 1
            num_comparisons += 1
    print(f"Number of comparisons: {num_comparisons}; number of swaps: {num_swaps}")
```

```python
## test
copied_list = li_master.copy()
comb_sort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```

## Insertion sort

```python
def insertion_sort(ls: list[int]) -> list[int]:
    n = len(ls)
    num_comparisons = 0
    for key_inx in range(1, n):
        key = ls[key_inx]
        first_less_inx = key_inx - 1
        while first_less_inx >= 0 and key < ls[first_less_inx]:
            num_comparisons += 1
            ls[first_less_inx+1] = ls[first_less_inx]
            first_less_inx -= 1
        ls[first_less_inx+1] = key
    print(f"Number of comparisons: {num_comparisons}")
```

```python
## test
copied_list = li_master.copy()
insertion_sort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```

## Mergesort

```python
def merge_sort(li):
    if len(li) <= 1:
        return li

    mid = len(li)//2

    left_sorted = merge_sort(li[:mid])
    right_sorted = merge_sort(li[mid:])

    return merge(left_sorted, right_sorted)


def merge(left, right):
    result = []

    # while both lists have elements
    while len(left) > 0 and len(right) > 0:
        if left[0] < right[0]:
            result.append(left[0])
            left = left[1:]
        else:
            result.append(right[0])
            right = right[1:]

    # At least one of left_sorted and right_sorted is now empty
    # Append the rest of the non-empty list to the result
    result = result + left
    result = result + right

    return result
```

```python
## test
copied_list = li_master.copy()
copied_list = merge_sort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```

## Quicksort

```python
def partition(li, inx1, inx2):
    pivot = li[inx1]
    low = inx1 + 1
    high = inx2
    while low < high:
        while low < len(li) and li[low] <= pivot:
            low += 1
        while high >= 0 and li[high] > pivot:
            high -= 1
        if high > low:
            li[high], li[low] = li[low], li[high]
    li[inx1], li[high] = li[high], li[inx1]
    return high

test_li = [7, 4, 3, 2, 14, 18, 1, 99, 6]
partition(test_li, 0, len(test_li)-1)
print(test_li)

test_li_2 = [100, 4, 3, 2, 14, 18, 1, 99, 6]
partition(test_li_2, 0, len(test_li_2)-1)
print(test_li_2)
            
            
```

```python
def quicksort(li, low=0, high=None):
    if high is None:
        high = len(li) - 1
    if low <= high:
        pivot_inx = partition(li, low, high)
        quicksort(li, low, pivot_inx - 1)
        quicksort(li, pivot_inx + 1, high)

```

```python
## test
copied_list = li_master.copy()
quicksort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```

## Heapsort

```python
def sift_down(li, i, n):
    """
    li: the list to sort
    i: first index in the unordered part of the list, which is a damaged heap
    n: first index in the ordered part of the list
    """
    largest = i
    left = 2*i + 1 ## index of left child
    right = 2*i + 2 ## index of right child

    if left < n and li[left] > li[largest]:
        largest = left
    if right < n and li[right] > li[largest]:
        largest = right
    
    ## largest is now the index of the largest of the top 3 elements
    ## in the damaged heap
    ## Note: we tested for left < n and right < n because if the heap
    ## is unbalanced, the root node might contain just 1 child or 0 children
    if largest != i:
        ## if necessary, replace the root with the true largest element
        li[i], li[largest] = li[largest], li[i]
        ## next, repeat the same, but 1 level down
        sift_down(li, largest, n)

def heapify(li):
    """
    Constructs the heap for the full list. Does this via sift_down method.
    More specifically: calls sift_down on the leaf nodes plus 1 parent.
    Then continue by adding all parents 1 by 1.
    """
    n = len(li)
    for i in range(n // 2 - 1, -1, -1):
        sift_down(li, i, n)

def heap_sort(li):
    """
    Step 1: construct the heap
    Step 2: in each iteration, swap the largest element on the heap with its last element
    Step 3: in each iteration, fix the damaged heap
    """
    n = len(li)

    heapify(li)

    for i in range(n-1, 0, -1):
        li[i], li[0] = li[0], li[i]
        sift_down(li, 0, i)
```

```python
## test
copied_list = li_master.copy()
heap_sort(copied_list)
print(f"original list: {li_master}")
print(f"sorted: {copied_list}")
```
