# Matrices as nested lists - transposal

```python
li = [[3, 7, 25, 1, 9], [2, 9, 8, 0, 55], [31, 8, 6, 1, 2]]
print(li)
```

```python
transposed = [[] for _ in range(len(li[0]))]
print(transposed)
```

```python
for col in range(len(li[0])):
    for row in range(len(li)):
        transposed[col].append(li[row][col])
print(transposed)
```

# Linked list

```python
class LinkedList:
    class Node:
        """
        Embedded class since Node is never needed outside of LinkedList
        """
        def __init__(self, value):
            self.value = value
            self.next = None
        
    def __init__(self):
        self.first_node = None

    def __str__(self):
        s = ""
        current = self.first_node
        while current is not None:
            s = s + f"{current.value}, "
            current = current.next
        return s

    def __iter__(self):
        """see use-case in print() method at the bottom"""
        current = self.first_node
        while current is not None:
            yield current.value
            current = current.next

    def append(self, value):
        node_to_append = self.Node(value)
        if self.first_node is None:
            self.first_node = node_to_append
        else:
            current = self.first_node
            while current.next is not None:
                current = current.next
            current.next = node_to_append

    def remove(self, value):
        previous, current = None, self.first_node
        while current is not None and current.value != value:
            previous = current
            current = current.next
        if current is not None:
            if previous is not None:
                previous.next = current.next ## even if this is None, no problem
            else:
                self.first_node = current.next ## even if this is None, no problem
            del current

    def print(self):
        values = [str(value) for value in self]
        print(f"Values in linked list: {values}")
```

```python
llist = LinkedList()
llist.append(5)
llist.append("hello")
llist.append(True)
llist.append(88)
llist.append("wicked cool!")
llist.append(3.14)
llist.print()
print("You can also print like this via __str__() method:")
print(llist)
```

```python
llist.print()
llist.remove(5)
llist.remove(True)
llist.remove(3.14)
llist.print()
```

# Doubly linked list

```python
class DoublyLinkedList:
    class Node:
        """
        Embedded class since Node is never needed outside of LinkedList
        """
        def __init__(self, value):
            self.value = value
            self.next = None
            self.prev = None
        
    def __init__(self):
        self.first_node = None
        self.last_node = None

    def __str__(self):
        s = "forward: "
        current = self.first_node
        while current is not None:
            s = s + f"{current.value}, "
            current = current.next
        s = s + "\nbackwards: "
        current = self.last_node
        while current is not None:
            s = s + f"{current.value}, "
            current = current.prev
        return s + "\n...\n"

    def __iter__(self):
        """see example use-case in print() method at the bottom"""
        current = self.first_node
        while current is not None:
            yield current.value
            current = current.next

    def reverse_iter(self):
        """see example use-case in print() method at the bottom"""
        current = self.last_node
        while current is not None:
            yield current.value
            current = current.prev

    def append(self, value):
        node_to_append = self.Node(value)
        if self.last_node is None:
            self.first_node = node_to_append
            self.last_node = node_to_append
        else:
            node_to_append.prev = self.last_node
            self.last_node.next = node_to_append
            self.last_node = node_to_append

    def prepend(self, value):
        node_to_prepend = self.Node(value)
        if self.first_node is None:
            self.first_node = node_to_prepend
            self.last_node = node_to_prepend
        else:
            self.first_node.prev = node_to_prepend
            node_to_prepend.next = self.first_node
            self.first_node = node_to_prepend

    def remove(self, value):
        current = self.first_node
        while current is not None and current.value != value:
            current = current.next
        if current is not None:
            ## if there is only 1 node and we are deleting it
            if current.next is None and current.prev is None:
                self.first_node = None
                self.last_node = None
            ## if there are at least 2 nodes and we are deleting the first one
            elif current.prev is None:
                self.first_node = current.next
                current.next.prev = None
            ## if there are at least 2 nodes and we are deleting the last one
            elif current.next is None:
                self.last_node = current.prev
                current.prev.next = None
            ## finally, the general case
            else:
                current.prev.next = current.next
                current.next.prev = current.prev
            del current

    def print(self, reverse=False):
        values = [str(value) for value in self] if not reverse else [str(value) for value in self.reverse_iter()]
        title = "Values in linked list in reverse" if reverse else "Values in linked list"
        print(f"{title}: [{', '.join(values)}]")
```

```python
llist = DoublyLinkedList()
llist.append("hello")
llist.append(True)
llist.append(88)
llist.append("wicked cool!")
llist.append(3.14)
llist.prepend(5)
llist.print()
llist.print(True)
print("You can also print like this via __str__() method:")
print(llist)
```

```python
llist.print()
llist.remove(5)
llist.remove(True)
llist.remove(3.14)
llist.print()
```
